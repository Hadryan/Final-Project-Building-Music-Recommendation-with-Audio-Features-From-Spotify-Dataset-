```python
import pandas as pd
```

```python
df = pd.read_csv('spotify_data.csv')
df_artists = pd.read_csv('my_top_artists.csv')
```

```python
df
```

```python
# dropping unnecessary column
df.drop(['Unnamed: 0', 'duration_ms', 'popularity', 'tempo', 'key', 'time_signature', 'loudness', 'mode'], axis = 1, inplace = True)
df
```

```python
df_filtered = df[df['artist_name'] == df_artists['artist_name'][0]]
for i in range(1, len(df_artists['artist_name'])):
    df_filtered = df_filtered.append(df[df['artist_name'] == df_artists['artist_name'][i]], ignore_index = True)
```

```python
df = df_filtered
```

```python
# checking nan value
df.isna().sum()
```

```python
# counting duplicates
grouped = df.groupby(['artist_name','track_name'], as_index=True).size()
grouped[grouped > 1].count()
```

```python
# dropping duplicates
df.drop_duplicates(subset=['artist_name','track_name'], inplace=True)
df.shape
```

```python
df.reset_index(inplace = True)
df.drop(['index'], axis = 1, inplace = True)
df
```

## Clustering

```python
from sklearn.cluster import KMeans
```

```python
from yellowbrick.cluster.elbow import kelbow_visualizer
```

```python
kelbow_visualizer(KMeans(), df.drop(['artist_name', 'track_name', 'track_id', 'uri'], axis = 1), k=(2,15))
```

```python
model = KMeans(n_clusters = 6)
model.fit(df.drop(['artist_name', 'track_name', 'track_id', 'uri'], axis = 1))
model.labels_
```

```python
plt.figure(figsize = (10,5))
sns.distplot(model.labels_, kde = False)
plt.savefig('label_dist.png')
```

```python
df['playlist_num'] = model.labels_ + 1
```
