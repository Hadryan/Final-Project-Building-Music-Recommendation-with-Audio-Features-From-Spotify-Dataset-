```python
import pandas as pd
```

```python
artist = pd.read_csv('my_top_artists.csv')
```

```python
import json
import requests
```

```python
related_artist = []
```

### Getting related artists based on user's top artists

```python
for i in artist['artist_id']:
    url = 'https://api.spotify.com/v1/artists/{}/related-artists'.format(i)
    headers = {'Authorization': "Bearer {}".format(TOKEN)}
    r = requests.get(url, headers=headers)
    parsed = json.loads(r.text)
    related_artist.append(parsed)
```

### Getting top track from related artists

```python
for i in range(0, len(artist['artist_id'])):
    for j in range(0,len(related_artist[i]['artists'])):
        url = 'https://api.spotify.com/v1/artists/{}/top-tracks?country=ID'.format(related_artist
                                                                                   [i]['artists'][j]['id'])
        headers = {'Authorization': "Bearer {}".format(TOKEN)}
        r = requests.get(url, headers=headers)
        parsed = json.loads(r.text)
        artist_top_track.append(parsed)
```

```python
artist_name = []
track_name = []
track_id = []
uri = []
```

```python
for i in range(0,len(artist_top_track)):
    for j in range(0, len(artist_top_track[i]['tracks'])):
        artist_name.append(artist_top_track[i]['tracks'][j]['artists'][0]['name'])
        track_name.append(artist_top_track[i]['tracks'][j]['name'])
        track_id.append(artist_top_track[i]['tracks'][j]['id'])
        uri.append(artist_top_track[i]['tracks'][j]['uri'])
```

```python
df_target = pd.DataFrame({'artist_name':artist_name,'track_name':track_name,'track_id':track_id,'uri':uri})
print(df_target.shape)
# check if everything is ok
df_target.head()
```

```python
# dropping duplicates
df_target.drop_duplicates(subset=['artist_name','track_name'], inplace=True)
df_target.shape
```

```python
df_target = df_target.reset_index()
```

```python
df_target.drop(['index'], axis = 1, inplace = True)
```

```python
df_target
```

### Getting the audio features of the tracks

```python
rows = []

for i in range(0,len(df_target['track_id'])):
    url = "https://api.spotify.com/v1/audio-features?ids={}".format(df_target['track_id'][i])
    headers = {'Authorization': "Bearer {}".format(TOKEN)}
    r = requests.get(url, headers=headers)
    parsed = json.loads(r.text)
    rows.append(parsed)
```

```python
df_audio_features = pd.DataFrame.from_dict(rows[0]['audio_features'],orient='columns') 
for i in range(1, len(rows)):
    df_audio_features = df_audio_features.append(pd.DataFrame.from_dict(rows[i]['audio_features'],orient='columns'), 
                                                 ignore_index=True)
```

```python
df_audio_features
```

```python
columns_to_drop = ['analysis_url','track_href','type']
df_audio_features.drop(columns_to_drop, axis=1,inplace=True)

df_audio_features.rename(columns={'id': 'track_id'}, inplace=True)

df_audio_features.shape
```

```python
df_audio_features.drop([0], axis=1,inplace=True)
```

```python
df_audio_features
```

```python
# merge both dataframes 
df = pd.merge(df_target, df_audio_features, on = 'track_id',how = 'inner') 
df.head()
```

```python
df.drop(['uri_y'], axis = 1, inplace = True)
df.rename(columns={'uri_x': 'uri'}, inplace=True)
df
```
